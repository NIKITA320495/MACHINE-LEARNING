```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df=df.drop(columns=['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'])
df
```

```python
inputs=df.drop('Survived',axis='columns')
target=df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
model=LabelEncoder()
```

```python
l_Pclass=LabelEncoder()
l_Sex=LabelEncoder()
l_Age=LabelEncoder() 
l_Fare=LabelEncoder()
```

```python
inputs['Pclass1']=l_Pclass.fit_transform(inputs['Pclass'])
inputs['Sex1']=l_Sex.fit_transform(inputs['Sex'])
inputs['Age1']=l_Age.fit_transform(inputs['Age'])
inputs['Fare1']=l_Fare.fit_transform(inputs['Fare'])
```

```python
inputs
```

```python
inputs=inputs.drop(['Pclass','Sex','Age','Fare'],axis='columns')
```

```python
print(inputs.head(10))
print(target.head(10))
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,target)
```
